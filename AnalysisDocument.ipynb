### Célula de preparação de todos os DF's a serem estudados.

```python
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
pd.set_option('display.max_rows', None)  # Mostra todas as linhas
pd.set_option('display.max_columns', None)  # Mostra todas as colunas

experiments = [
    pd.read_csv("Experimento-1.csv"),
    pd.read_csv("Experimento-2.csv"),
    pd.read_csv("Experimento-3.csv"),
    pd.read_csv("Experimento-4.csv"),
    pd.read_csv("Experimento-5.csv")
]
experiments[1].Heuristic.fillna("None",inplace = True)
df_infos = []
for exp in experiments:
    # Verifica se a coluna 'Heuristic' existe no DataFrame
    if 'Heuristic' in exp.columns:
        grouping_columns = ['Algorithm', 'Cost Function', 'Heuristic']
    else:
        grouping_columns = ['Algorithm', 'Cost Function']
    
    # Faz o agrupamento com base nas colunas disponíveis
    grouped = exp.groupby(grouping_columns)[['Path Cost', 'Generated Nodes', 'Visited Nodes']].agg(
        ['mean', 'median', 'std']
    )
    
    # Adiciona o resultado à lista
    df_infos.append(grouped.reset_index())
```
